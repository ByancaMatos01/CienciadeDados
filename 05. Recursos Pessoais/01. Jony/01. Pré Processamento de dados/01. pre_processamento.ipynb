### Import da base

```python
import pandas as pd
import numpy as np

url = r'C:\Users\joth1\Desktop\byanca\CienciadeDados\05. Recursos Pessoais\01. Jony\01. Pré Processamento de dados\heart_tratado (1).csv'

df = pd.read_csv(url, sep=';', encoding='utf-8')
df
```

### Transformação do dataframe em matriz

```python
previsores = df.iloc[:, 0:11].values
previsores.shape
```

### Aplicação do principio do LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

lista = [1,2,6,8,10]
for i in lista:
    previsores[:, i] = LabelEncoder().fit_transform(previsores[:,1])
previsores
```

### Aplicação do OneHotEncoder

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

previsores2 = df.iloc[:, 0:11].values
previsores2.shape


previsores3 = ColumnTransformer(
    transformers=[('OneHot'
                   , OneHotEncoder()
                   , [1, 2, 6, 8, 10])]
    ,remainder='passthrough'
).fit_transform(previsores2)


display(previsores3)
previsores3.shape
```

### Apliação do Principio de escalonamentos   
#### StandardScaler

```python
from sklearn.preprocessing import StandardScaler
previsores4 = StandardScaler().fit_transform(previsores3)
```

```python
df = pd.DataFrame(previsores4)
display(df)
```
